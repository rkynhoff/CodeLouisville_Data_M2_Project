```python
#ADDED to V2
#Cleaning Data - Drop duplicate rows 
laptop_data_deduped = laptop_data.drop_duplicates()
display(laptop_data_deduped)
```

```python
#ADDED to V2
#counts the total number of "?" (or question marks [qm] by column 
qm_counts = (laptop_data_deduped == "?").sum()
#sums the total number of qms in the df
total_qms = qm_counts.sum()
print("Counts of '?' by column: ") 
print(qm_counts)
print()
print(f"Total count of '?' in the df: {total_qms}.")
```

```python
#ADDED to V2
#Identify index # of row with "?" to use for dropping these rows in future code
questmark1 = laptop_data_deduped[laptop_data_deduped.Inches == "?"]
questmark2 = laptop_data_deduped[laptop_data_deduped.Memory == "?"]
questmark3 = laptop_data_deduped[laptop_data_deduped.Weight == "?"]
print(questmark1.index)
print(questmark2.index)
print(questmark3.index)
```

```python
#drop rows with incomplete data (a.k.a. rows containing "?" in place of column value)
laptop_data_drop_rows = laptop_data_deduped.drop(labels = [208, 476, 770], axis = 0)
laptop_data_drop_rows
```

```python
# ADDED to to Data Type Adjusments V2
laptop_data_drop_rows["Weight"] = laptop_data_drop_rows["Weight"].str.replace("kg","")
laptop_data_drop_rows.head(5)
```

```python
# Added to Data Type Adjusments V2
laptop_data_drop_rows[["Weight","Inches"]]=laptop_data_drop_rows[["Weight","Inches"]].astype(float)
laptop_data_drop_rows.head(5) #removed this line
```

```python
# Added to line above in Data Type Adjustments V2 as check
laptop_data_drop_rows.info()
```

```python
#Added to V2
#Cleaning Data - Drop Columns & display column info to ensure correct column dropped
laptop_data_deduped_dropcol = laptop_data_drop_rows.drop("Unnamed: 0",axis=1)
laptop_data_deduped_dropcol.columns
```

```python
#Cleaning Data - Update price column name to 'Price_RUP', round 'Price_RUP' column to 2 decimals & display column info to ensure correct column renamed
#Added to V2 plus several other columns
laptop_data_deduped_dropcol.rename(columns={"Price": "Price_RUP"}, inplace=True)
laptop_data_deduped_dropcol.columns
```

```python
#Cleaning Data - Round "Price_Rub" column to two decimal places & Display top rows to ensure correct formatting
laptop_data_prices = laptop_data_deduped_dropcol.round({"Price_RUP":2})
laptop_data_prices.head()
```

```python
#Cleaning Data - Add a new price column converting RUP to USD using latest exchange rate of $1 to 0.12 RUP as of 12/05/2024 and name new column 'Price_USD'
laptop_data_prices["Price_USD"] = (laptop_data_prices["Price_RUP"] * 0.012).round(2)
laptop_data_prices.head()
```
