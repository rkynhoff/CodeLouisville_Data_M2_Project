### **Selection Options**

```python
# Drop columns that won't be used
laptop_data_drop = laptop_data.drop("Unnamed: 0", axis = 1)
laptop_data_drop 
```

## **Data Cleaning & Preparation**

### **Handle Duplicated Rows**

```python
# Display count of duplicated rows
laptop_data_drop.duplicated().value_counts()
```

```python
# Determine how many of the 58 rows are rows with null values
# These will get dropped in subsequent code along with other duplicated rows
laptop_data_drop.isnull().any(axis=1).sum()
```

```python
# Displays only rows with null values
laptop_data_drop[laptop_data_drop.isna().any(axis=1)]
```

```python
# Drop duplicated rows
laptop_data_deduped = laptop_data_drop.drop_duplicates()
# Then display count of duplicated rows
laptop_data_deduped.duplicated().value_counts()
```

### **Handle Missing Values**

```python
# Boolean output confirming if there are still rows with null values
laptop_data_deduped.isnull().values.any()
# Output should be 'np.True_'
```

```python
# Row count of null data
laptop_data_deduped.isnull().any(axis = 1).sum()
```

```python
# Displays only rows with null values 
laptop_data_deduped[laptop_data_deduped.isna().any(axis=1)]
```

```python
# Drop remaining rows with null values
laptop_data_deduped_drop = laptop_data_deduped.dropna()
```

```python
# Check to ensure all null values have been removed
laptop_data_deduped_drop.duplicated().value_counts()
```

### **Handle Special Characters**

```python
# Identify if special character exits, i.e. "?"
# Counts total # of "?" by column 
qm_counts = (laptop_data_deduped_drop == "?").sum()
# Sums total # of "?" in the df by column
total_qms = qm_counts.sum()
print("Counts of '?' by column: ")
print()
print(qm_counts)
print()
print(f"Total count of '?' in the df: {total_qms}.")
```

```python
# Identify index # of row with "?" to use for dropping these rows as inconclusive data will not aid in my decision
questmark1 = laptop_data_deduped_drop[laptop_data_deduped_drop.Inches == "?"]
questmark2 = laptop_data_deduped_drop[laptop_data_deduped_drop.Memory == "?"]
questmark3 = laptop_data_deduped_drop[laptop_data_deduped_drop.Weight == "?"]
print(questmark1.index)
print(questmark2.index)
print(questmark3.index)
```

```python
# Drop rows with inconclusive/incomplete data (a.k.a. rows containing "?" in place of null or column value)
laptop_data_drop_rows = laptop_data_deduped_drop.drop(index = [208, 476, 770], axis = 0)
laptop_data_drop_rows
```

```python
# Rechecking total count of "?" in df to ensure all rows containing the value were dropped
qm_counts = (laptop_data_drop_rows == "?").sum()
# Sums total # of "?" in the df by column
total_qms = qm_counts.sum()
print(f"Total count of '?' in the df: {total_qms}.")
```

### **Data Type Adjustments**

```python
# Remove "kg" from data in Weight column in preparation to 
laptop_data_drop_rows["Weight"] = laptop_data_drop_rows["Weight"].str.replace("kg","")
# Display first three rows to ensure "kg" was removed from Weight column
laptop_data_drop_rows.head(3)
```
